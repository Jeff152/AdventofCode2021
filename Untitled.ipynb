```python
#Day 1
with open('day1','r') as f:
    values = f.read().splitlines()
count = 0
values = [int(v) for v in values]
temp = values[0]
for v in values:
    if int(v) > temp:
        count += 1
    temp = v
print(f'Day 1 Part 1 Answer\n{count}')
count = 0
otherSum = []
for i in range(0,len(values)-2):
    otherSum.append(values[i]+values[i+1]+values[i+2])
temp = otherSum[0]
for v in otherSum:
    if v > temp:
        count +=1
    temp = v
    
print(f'Day 1 Part 2 Answer\n{count}')
```

```python
#Day 2
total = {'forward': 0, 'up':0,  'down':0}
with open('day2', 'r') as f:
    line = f.readline()
    while line:
        direction, value = line.split()
        total[direction] += int(value)
        line = f.readline()
ans = total['forward'] * (total['down']-total['up'])
print(f'Day 2 Part 1 Answer\n{ans}')
with open('day2', 'r') as f:
    line = f.readline()
    aim = 0
    posx = 0
    posy = 0
    while line:
        direction, value = line.split()
        value = int(value)
        if direction == 'forward':
            posx += value
            posy += aim * value
        elif direction == 'up':
            aim -= value
        elif direction == 'down':
            aim += value
        line = f.readline()
    print(f'Day 2 Part 2 Answer\n{posx*posy}')
            
```

```python
#Day 3
mostcommon = {index:0 for index in range(0,12)}
values = []
with open('day3', 'r') as f:
    line = f.readline()
    while line:
        values.append(line)
        
        for i,bit in enumerate(line):
            if bit == '0':
                mostcommon[i] -= 1
            elif bit == '1':
                mostcommon[i] += 1
        line = f.readline()
gamma = ''
epsilon = ''
for key, value in mostcommon.items():

    if value < 0:
        gamma += '0'
        epsilon += '1'
    else:
        gamma += '1'
        epsilon += '0'
print(f'Gamma:{gamma} -> {int(gamma,2)}')
print(f'Epsilon:{epsilon} -> {int(epsilon,2)}')
print(f'Day 3 Part 1 Answer\n{int(gamma,2)*int(epsilon,2)}')
#Part 2
#Finding the Oxygen Generator Rating and  (Most Common)
oxy, c02 = values, values
for i in range(0,12):
    mostCommon = 0
    for bit in oxy:
        if bit[i] == '0':
            mostCommon -= 1
        else:
            mostCommon += 1
    if mostCommon < 0:
        oxy = [x for x in oxy if x[i] != '1']
    else:
        oxy = [x for x in oxy if x[i] != '0']
    if len(oxy) == 1:
        oxyRating = oxy[0]
        continue
for i in range(0,12):
    mostCommon = 0
    for bit in c02:
        if bit[i] == '0':
            mostCommon -= 1
        else:
            mostCommon += 1
    if mostCommon < 0:
        c02 = [x for x in c02 if x[i] != '0']
    else:
        c02 = [x for x in c02 if x[i] != '1']
    if len(c02) == 1:
        c02Rating = c02[0]
        continue

            
        
print(f'Day 3 Part 2 Answer')
print(f'Oxygen Rating: {oxyRating} -> {int(oxyRating,2)}')
print(f'C02 Rating: {c02Rating} -> {int(c02Rating,2)}')
print(f'{int(oxyRating,2)*int(c02Rating,2)}')
```

```python
#Day 4
with open('day4','r') as f:
    solution = [0 for x in range(0,12)]
    line = f.readline()
    guesses = line.split(',')
    line = f.readline()
    line = f.readline()
    boards = []
    board = []
    solutions = []
    while line:
        #Create new board
        if line == '\n':
            boards.append(board)
            solutions.append(solution)
            solution = [0 for x in range(0,12)]
            board = []
            line = f.readline()
            continue
        board.append(line.split())
        line = f.readline()
def doGuess(guesses, boards, solutions):
    
    for guess in guesses:
        guess = int(guess)
        
        #Checking all the boards if the draw is in the board
        for i, board in enumerate(boards):
            for y, row in enumerate(board):
                for x, value in enumerate(row):
                    if guess == int(value):
                        solutions[i][x] += 1
                        solutions[i][y+5] += 1
                        if x == y:
                            solutions[i][10] += 1
                        if x + y == 4:
                            solutions[i][11] += 1
                    for s in solutions[i]:
                        if s == 5:
                            return guess, board
                    
    
lastGuess, solutionBoard = doGuess(guesses, boards, solutions)
draws = guesses[:guesses.index(str(lastGuess))+1]
unmarkedSum = 0
for bleh in solutionBoard:
    for value in bleh:
        if value not in draws:
            unmarkedSum += int(value)
ans = unmarkedSum * lastGuess
print(f'Day 4 Part 1 Answer')
print(f'Final Bingo Score: {ans}')

with open('day4','r') as f:
    solution = [0 for x in range(0,12)]
    line = f.readline()
    guesses = line.split(',')
    line = f.readline()
    line = f.readline()
    boards = []
    board = []
    solutions = []
    while line:
        #Create new board
        if line == '\n':
            boards.append(board)
            solutions.append(solution)
            solution = [0 for x in range(0,12)]
            board = []
            line = f.readline()
            continue
        board.append(line.split())
        line = f.readline()
        
def lastBoard(guesses, boards, solutions):
    solutionBoard = None

    for guess in guesses:
        solved = 0
        guess = int(guess)
        
        #Checking all the boards if the draw is in the board
        for i, board in enumerate(boards):
            for y, row in enumerate(board):
                for x, value in enumerate(row):
                    if guess == int(value):
                        solutions[i][x] += 1
                        solutions[i][y+5] += 1
                        if x == y:
                            solutions[i][10] += 1
                        if x + y == 4:
                            solutions[i][11] += 1
            for s in solutions[i]:
                if s > 5:
                    print(f's: {s} i: {i}')
                if s >= 5:
                    solved += 1
                    break
        if len(boards) == solved:
            return guess, solutionBoard
        if len(boards) - solved == 1:
            for i, solution in enumerate(solutions):
                weh = False
                for s in solution:
                    if s >= 5:
                        weh = True
                if not weh:
                    solutionBoard = boards[i]
lastGuess, solutionBoard = lastBoard(guesses, boards, solutions)
draws = guesses[:guesses.index(str(lastGuess))+1]
unmarkedSum = 0
for bleh in solutionBoard:
    for value in bleh:
        if value not in draws:
            unmarkedSum += int(value)
ans = unmarkedSum * lastGuess
print(f'Day 4 Part 2 Answer')
print(f'Final Bingo Score: {ans}')

                    
                    
                    
```

```python
#Day 5 Part 1
with open('day5', 'r') as f:
    points = {}
    line = f.readline()
    while line:
        point1, _ , point2 = line.split()
        point1 = [int(x) for x in point1.split(',')]
        point2 = [int(x) for x in point2.split(',')]
        if point1[0] != point2[0] and point1[1] != point2[1]:
            line = f.readline()
            continue
        if point1[0] > point2[0]:
            greaterX = point1[0]
            lowerX = point2[0]
        else:
            greaterX = point2[0]
            lowerX = point1[0]
        if point1[1] > point2[1]:
            greaterY = point1[1]
            lowerY = point2[1]
        else:
            greaterY = point2[1]
            lowerY = point1[1]
            
        for x in range(lowerX,greaterX+1):
            for y in range(lowerY,greaterY+1):
                point = f'({x},{y})'
                if point not in points:
                    points[point] = 0
                points[point] += 1
        line = f.readline()
    ans = 0
    for v in points.values():
        if v > 1:
            ans += 1
print(f'Day 5 Part 1 Answer')
print(f'Number of Points Overlapped: {ans}')
            
#Day 5 Part 1
with open('day5test', 'r') as f:
    points = {}
    line = f.readline()
    while line:
        point1, _ , point2 = line.split()
        point1 = [int(x) for x in point1.split(',')]
        point2 = [int(x) for x in point2.split(',')]
        if point1[0] > point2[0]:
            greaterX = point1[0]
            lowerX = point2[0]
        else:
            greaterX = point2[0]
            lowerX = point1[0]
        if point1[1] > point2[1]:
            greaterY = point1[1]
            lowerY = point2[1]
        else:
            greaterY = point2[1]
            lowerY = point1[1]
            
        # Diagonal Line
        if point1[0] != point2[0] and point1[1] != point2[1]:
            print(line)
            while lowerX != greaterX+1:
                point = f'({lowerX},{lowerY})'
                if point not in points:
                    points[point] = 0
                points[point] += 1
                lowerX += 1
                lowerY += 1
            line = f.readline()
            continue
            
        for x in range(lowerX,greaterX+1):
            for y in range(lowerY,greaterY+1):
                point = f'({x},{y})'
                if point not in points:
                    points[point] = 0
                points[point] += 1
        line = f.readline()
    ans = 0
    for v in points.values():
        if v > 1:
            ans += 1
print(f'Day 5 Part 1 Answer')
print(f'Number of Points Overlapped: {ans}')
            
```
